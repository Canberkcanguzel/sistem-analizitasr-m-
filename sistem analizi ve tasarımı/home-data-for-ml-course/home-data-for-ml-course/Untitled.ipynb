```python
%matplotlib inline
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns

y = train['SalePrice']

plt.figure(2)  # Grafik için yeni bir figür oluştur
plt.title('Normal')  # Başlık ekle

# Histogramı çiz ve normal dağılımla uyumlu hale getir
sns.histplot(y, kde=False, stat="density", bins=30, color='blue')  # Histogram çiz

# Fit edilen normal dağılım eğrisini çiz
xmin, xmax = plt.xlim()  # Grafik sınırlarını al
x = np.linspace(xmin, xmax, 100)  # Verinin aralığında değerler oluştur
p = st.norm.pdf(x, y.mean(), y.std())  # Normal dağılım fonksiyonu
plt.plot(x, p, 'k', linewidth=2)  # Normal dağılım eğrisini çiz
plt.show()
```

```python
test_normality = lambda x: st.shapiro(x.fillna(0))[1] < 0.01  # Şapironun p-değerini kontrol et

normal = pd.DataFrame(train[cuantitativo])  # Cuantitativo sütunlarını seç
normal = normal.apply(test_normality)  # test_normality fonksiyonunu her sütuna uygula

print(normal.any())  # Her sütun için normalite testinin sonucunu kontrol et
```

```python

cuantrain = train[['OverallQual','TotalBsmtSF','1stFlrSF','SalePrice','GarageCars','GarageArea', 'GrLivArea']]
cuantrain.hist(bins=30, figsize=(15,10))
plt.show()  # grafiklerin gösterilmesini garanti altına alır
```

```python
from sklearn import preprocessing

cols = cuantrain.columns
min_max_scaler = preprocessing.MinMaxScaler()
cuantrain_scaled = min_max_scaler.fit_transform(cuantrain)
cuantrain = pd.DataFrame(cuantrain_scaled, columns=cols)

cuantrain.hist(bins=30, figsize=(15,10))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model  import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataTrain, dataTest, dataTrain_price, dataTest_price = train_test_split(
    cuantrain[['OverallQual', '1stFlrSF', 'GarageArea', 'GrLivArea']],
    cuantrain['SalePrice'],
    test_size=0.2,
    shuffle=True,
    random_state=100
)
reg = LinearRegression().fit(dataTrain,dataTrain_price)
reg.score(dataTrain,dataTrain_price)
```

```python
from sklearn import metrics

# Tahmin yap
y_predict_test = reg.predict(dataTest)

# Hata metriği hesapla
mse = metrics.mean_squared_error(dataTest_price, y_predict_test)
print(f"MSE: {mse:.2f}")
```

```python
test.fillna(0,inplace=True)
test['SalePrice'] = reg.predict(test[['OverallQual', '1stFlrSF', 'GarageArea', 'GrLivArea']])
print(test['SalePrice'])
test[['Id','SalePrice']].to_csv('submussion.csv',index=False)
```
